# Analiza podatkov s pandas

```python
import pandas as pd
```

```python
artikli1 = pd.read_csv("artikli1.csv")
```

```python
artikli2 = pd.read_csv("artikli2.csv")
```

```python
artikli3 = pd.read_csv("artikli3.csv")
```

```python
artikli4 = pd.read_csv("artikli4.csv")
```

```python
artikli5 = pd.read_csv("artikli5.csv")
```

```python
#zdruzitev prve in druge csv datoteke
zdruzeno1 = artikli1.merge(artikli2, left_index=True, right_index=True)
```

```python
print(zdruzeno1)
```

```python
#zdruzitev prve, druge in tretje csv datoteke
zdruzeno2 = zdruzeno1.merge(artikli3, left_index=True, right_index=True)
```

```python
print(zdruzeno2)
```

```python
#zdruzitev prve, druge, tretje in cetrte csv datoteke
zdruzeno3 = zdruzeno2.merge(artikli4, left_index=True, right_index=True)
```

```python
print(zdruzeno3)
```

```python
#zdruzitev vseh csv datotek
zdruzeno4 = zdruzeno3.merge(artikli5, left_index=True, right_index=True)
```

```python
print(zdruzeno4)
```

```python
#prokaz nove tabele (prvih 19 artiklov in njihovih podatkov)
zdruzeno4.head(20)
```

```python
#sortirano po cenah (najcenejših 19)
zdruzeno4.sort_values('cena').head(20)
```

```python
#sortirano po cenah (najdražjih 19)
zdruzeno4.sort_values('cena').tail(20)
```

```python
zdruzeno3.head(20)
```

```python
zdruzeno3.groupby("datum")["cena"].mean().sort_values().plot.bar()
```

```python
#tortni prikaz kako so porazdeljene povprečne cene po datumih
zdruzeno3.groupby("datum")["cena"].mean().sort_values().plot.pie(autopct='%1.1f%%')
```

```python
zdruzeno3.groupby("datum")["cena"].mean()
```

```python
zdruzeno3.groupby("datum")["cena"].mean().sort_values()
```

```python
zdruzeno2.groupby("lokacija")["cena"].mean().sort_values().plot.bar()
```

```python
zdruzeno2.groupby("lokacija")["cena"].mean().sort_values()
```
